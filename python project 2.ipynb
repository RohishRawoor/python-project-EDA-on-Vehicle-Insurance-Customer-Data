```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1.i
table1=pd.read_csv('C:\\Users\\rohis\\Desktop\\skill lync\\python\\project 2\\Data to be used\\customer_details.csv')
```

```python
table2=pd.read_csv('C:\\Users\\rohis\\Desktop\\skill lync\\python\\project 2\\Data to be used\\customer_policy_details.csv')
```

```python
#1.ii
col1=['customer_id','Gender','age','driving licence present','region code','previously insured','vehicle age','vehicle damage']
col2=['customer_id','annual premium (in Rs)','sales channel code','vintage','response']
table1.columns=col1
table2.columns=col2

```

```python
#2.i

print(table1.isnull().sum())
print(table2.isnull().sum())
```

```python
table1=table1.dropna(subset='customer_id')
table2=table2.dropna(subset='customer_id')

#replacing numerical values with MEAN
table1[['age','driving licence present','region code']]=table1[['age','driving licence present','region code']].fillna(table1[['age','driving licence present','region code']].mean())
table2[['annual premium (in Rs)','sales channel code','vintage']]=table2[['annual premium (in Rs)','sales channel code','vintage']].fillna(table2[['annual premium (in Rs)','sales channel code','vintage']].mean())


#replacing categorical values with numerical values
table1['Gender']=table1['Gender'].replace(['Male','Female'],[1,0])
table1['vehicle age']=table1['vehicle age'].replace(['< 1 Year','1-2 Year','> 2 Years'],[0,1,2])
table1['vehicle damage']=table1['vehicle damage'].replace(['Yes','No'],[1,0])
```

```python
#remaining null values are caused by categorical values being changed to numbers
print(table1.isnull().sum())
print(table2.isnull().sum())
```

```python
#z_score method to find outliers
#z-score
outlier=[]
def detect_outlier(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
```

```python
#finding outliers column wise(for numerical) 
detect_outlier(table1['age'])
#NO outliers found
```

```python
detect_outlier(table2['vintage'])
#NO outliers found
```
